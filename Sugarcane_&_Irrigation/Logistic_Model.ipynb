```python
import pandas as pd
import joblib
import numpy as np

data = joblib.load("/content/drive/MyDrive/lr_model.pkl")
test= pd.read_csv('/content/drive/MyDrive/Irrigarion/test.csv')


model = data["model"]
feature_names = data["features"]

def predict_irrigation(input_data):

    if isinstance(input_data, pd.Series):
        new_df = input_data.to_frame().T
    else:
        new_df = pd.DataFrame([input_data])

    new_df = new_df.reindex(columns=feature_names, fill_value=0)

    probs = model.predict_proba(new_df)[0]

    classes = model.classes_

    pred_class = classes[np.argmax(probs)]

    print(pred_class)

    return {
        "class": pred_class,
        "probabilities": probs
    }
```

```python
row = test.iloc[0]
result = predict_irrigation(row)
```
