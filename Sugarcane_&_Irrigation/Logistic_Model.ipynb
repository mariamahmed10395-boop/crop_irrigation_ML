```python
import pandas as pd
import joblib
import numpy as np

data = joblib.load("/content/drive/MyDrive/lr_model.pkl")
test = pd.read_csv('/content/drive/MyDrive/Irrigarion/test.csv')

model = data["model"]
feature_names = data["features"]


test_aligned = test.reindex(columns=feature_names, fill_value=0)
probs = model.predict_proba(test_aligned)
preds = model.classes_[np.argmax(probs, axis=1)]

submission = pd.read_csv('/content/drive/MyDrive/Irrigarion/sample_submission.csv')
submission.iloc[:, 1] = preds

submission.to_csv("/content/drive/MyDrive/Irrigarion/submission.csv", index=False)
```
